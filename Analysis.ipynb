#### What products are most often sold together?

```python
all_data_cleaned.head()
```

```python
df = all_data_cleaned[all_data_cleaned['Order ID'].duplicated(keep=False)]
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df= df[['Order ID', 'Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter
count = Counter()
for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

count.most_common(10)
```

#### Which Products were sold the most

```python
all_data_cleaned.head()
```

```python
product_group = all_data_cleaned.groupby('Product')
product_group['Quantity Ordered'].sum()
```

```python
quantity_ordered= product_group['Quantity Ordered'].sum()
products = [product for product, df in product_group]
plt.bar(products, quantity_ordered)
plt.ylabel('Quantity Ordered')
plt.xlabel('Products')
plt.xticks(products, rotation ='vertical')
plt.show()
```

```python
prices = all_data_cleaned.groupby('Product')['Price Each'].mean()
print(prices)
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered)
ax2.plot(products, prices, 'b-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color = 'y')
ax2.set_ylabel('Price', color = 'b')
ax1.set_xticklabels(products, rotation = 'vertical')
plt.show()
```
